```python
# Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading csv file with pandas
df = pd.read_csv('C:\\Users\\hp\\Downloads\\householdtask3.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Scatter Plot between year and income
im = plt.scatter(df['year'],df['income'])
# adding title
plt.title('Scatter Plot')
# adding X, Y- axis Labels
plt.xlabel('Year')
plt.ylabel('Income')
# adding more customization
plt.grid(True)
plt.colorbar()
plt.show()
```

```python
# Line Chart between income and expenditure
plt.plot(df['income'], 'g', label = 'income')
plt.plot(df['expenditure'], 'r', label = 'expenditure')
# adding title
plt.title('Line Plot')
# adding X, Y- axis Labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# adding more customization
plt.grid(True)
plt.legend() 
plt.show()
```

```python
# Bar Chart between year and income
plt.bar(df['year'], df['income'])
# adding title
plt.title('Bar Plot')
# adding X, Y- axis Labels
plt.xlabel('Year')
plt.ylabel('Income')
plt.grid(True)
plt.show()
```

```python
# Histogram
plt.hist(df['income'], bins=20, color='purple')
# adding title
plt.title('Histogram')
# adding X, Y- axis Labels
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
